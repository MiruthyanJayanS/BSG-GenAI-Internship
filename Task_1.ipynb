```python
import pandas as pd
```

```python
df = pd.read_csv('E:\BCG\BCG_Data.csv')
```

```python
df.head()
```

```python
df = df.drop(columns=['Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10'])
```

```python
df
```

```python
df['Revenue Growth (%)'] = df.groupby(['Company'])['Total Revenue'].pct_change() * 100
df['Net Income Growth (%)'] = df.groupby(['Company'])['Net Income'].pct_change() * 100
```

```python
df
```

```python
df.fillna(0, inplace=True)
```

```python
df
```

```python
summary = df.groupby('Company').agg({
    'Revenue Growth (%)': 'mean',
    'Net Income Growth (%)': 'mean'
}).reset_index()
```
